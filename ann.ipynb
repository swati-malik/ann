```
y = data.iloc[:,13].values
```

```

from sklearn.model_selection import train_test_split
```

```
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```
le1 = LabelEncoder()
le2 = LabelEncoder()
```

```
x[:, 1]= le1.fit_transform(x[:, 1])
```

```
x[:,2] = le2.fit_transform(x[:,2])
```

```
x
```

```
#onehotencoding
onehotencoder = OneHotEncoder(categorical_features=[1])

```

```
x= onehotencoder.fit_transform(x).toarray()
```

```
x = pd.DataFrame(x)
```

```
x
```

```
x = x.iloc[:,1:]
```

```
x
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

feature scaling


```
from sklearn.preprocessing import StandardScaler
```

```

sc = StandardScaler()
```

```
x_train = sc.fit_transform(x_train)
```

```
x_test = sc.fit_transform(x_test)
```

```
import keras
```

```
from keras.models  import Sequential
```

```
from keras.layers import Dense
```

```
classifier = Sequential()
```

**for hidden layer **


-output_dim argument signifies the no. of nodes in the hidden layer and on experimental basis it can be calculated by taking the average of input and output nodes.

-init argument is used to initialize the weights.

-activation function for hidden layer is usually taken as relu.

-activation fucnction for output layer is taken 'sigmoid' if we have binary output and 'softmax ' if the output contains more than two categories.

```
#second hidden layer
classifier.add(Dense(output_dim=6,init = 'uniform',activation='relu',input_dim = 11))
```

```
#second hidden layer
classifier.add(Dense(output_dim=6,init = 'uniform',activation='relu'))
```

```
#output layer
classifier.add(Dense(output_dim=1,init = 'uniform',activation='sigmoid'))
```

```
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
classifier.fit(x_train,y_train,batch_size = 10,epochs = 100)
```

```
y_pred = classifier.predict(x_test)
```

```
y_pred = y_pred>0.5
```

```
from sklearn.metrics import confusion_matrix
```

```
cm = confusion_matrix(y_test,y_pred)
```
